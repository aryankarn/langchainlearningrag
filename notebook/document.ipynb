```python
## Create a simple txt file
import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_texts = {
    "../data/text_files/python_intro.txt": """
    Python is a beginner-friendly, high-level programming language used for web development,
    data analysis, automation, artificial intelligence, machine learning, and scripting.

    One of Python's biggest strengths is its clear and readable syntax. Programs written in
    Python often look close to plain English, which helps new programmers focus on solving
    problems instead of struggling with complex language rules.

    Python also has a large ecosystem of libraries and frameworks. Tools like NumPy, pandas,
    FastAPI, Django, LangChain, and PyTorch allow developers to build real-world applications
    quickly without writing everything from scratch.

    Because Python is widely used in education, startups, research, and large companies, it is
    a practical language for both beginners and experienced developers.
    """,
}

for filepath, content in sample_texts.items():
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)
print("Sample text files created successfully.")
```

```python
## Textloader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")
document = loader.load()

print(document)
```

```python
## Directory Loader

from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader("../data/text_files", glob="*.txt", loader_cls=TextLoader, loader_kwargs={"encoding": "utf-8"})
###glob in the above line is the pattern to match the files in the directory. In this case, it will load all .txt files in the specified directory.
documents = loader.load()

print(documents)
```

```python
### pdf loader

from langchain_community.document_loaders import DirectoryLoader, PyMuPDFLoader

loader = DirectoryLoader("../data/pdf", glob="*.pdf", loader_cls=PyMuPDFLoader)
documents = loader.load()

documents
```
